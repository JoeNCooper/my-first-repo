Display the top 5 video games by global sale

```python
grouped_df = df.groupby('Name')['Global_Sales'].nlargest(5)
print(grouped_df)
```

Calculate the total sales in North America for the 'Action' genre.

```python
df.pivot_table[index='NA_Sales',columns='Genre',values='Action']
```

Find the top 3 publishers by global sales.

```python
grouped_df = df.groupby('Publisher')['Global_Sales'].sum().nlargest(3)
print(grouped_df)
```

Calculate the average sales in Europe for the 'Shooter' genre.

```python
grouped_df = df.groupby('Shooter')['EU_Sales'].mean()
print(grouped_df)
```

Find the game with the highest sales in Japan in the 'Sports' genre

```python
grouped_df = df.groupby('Genre')['JP_Sales'].transform(max)
print(grouped_df)
```
